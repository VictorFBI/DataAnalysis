```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency as chi2

df = pd.read_csv('/content/smoking_driking_dataset_Ver01.csv', sep=',')
features = ['sex', 'age', 'height', 'weight', 'waistline', 'sight_left', 'sight_right', 'hear_left', 'hear_right', 'SBP', 'DBP', 'BLDS', 'tot_chole', 'hemoglobin', 'urine_protein', 'SMK_stat_type_cd', 'DRK_YN']
df = df[features]
df.sample(n=10)
```

```python
shape = df.shape
print("Число наблюдений = {}".format(shape[0]))
print("Число признаков = {}".format(shape[1]))
print(f"Число пустых значений в датасете = {df.isna().sum().sum()}")
```

```python
df.dtypes
```

Заметим, что все признаки принимают логичный тип данных, так что менять ничего не будем



Проверим уникальность значений на категориальных переменных

```python
for feature in ['sex', 'hear_left', 'hear_right', 'SMK_stat_type_cd', 'DRK_YN']:
  print(np.unique(df[feature]))
```

По имеющимся данным рассчитаем индекс массы тела (BMI - Body Mass Index)  
$BMI = \frac{weight}{height^2}$

```python
df['BMI'] = df['weight'] / (df['height'] ** 2)
df.sample(n=10)
```

Сделаем категориальную переменную на основе переменной возраст (age). Правило составления будет таким:  
до 18 лет - young  
18-55 - adult  
55+ - pensioner  

```python
def decide_age_group(age):
  if age < 18:
    return 'young'
  if age > 55:
    return 'pensioner'
  return 'adult'

df['age_group'] = df['age'].apply(decide_age_group)
df.sample(n=10)
```

Перейдем к визуализации данных

```python
sns.countplot(x='sex', hue='DRK_YN', data=df, palette='Set2')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.legend(title='Drink or not')
plt.title('Distribution of drinkers and non-drinkers over sexes', fontweight='bold', fontsize=18)
plt.show()
```

График показывает, что пьющих мужчин гораздо больше, чем пьющих женщин

```python
sns.countplot(x='sex', hue='SMK_stat_type_cd', data=df, palette='Set2')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.legend(title='Smoke status')
plt.title('Distribution of smoke statuses over sexes', fontweight='bold', fontsize=18)
plt.show()
```

Данный график показывает, что среди мужчин наблюдается равномерное распределение по статусу курения, в то время как по женщинам бОльшая часть никогда не курила

```python
fig, axs = plt.subplots(1, 2, figsize=(16, 6))

sns.histplot(x='hemoglobin', data=df[df['sex'] == 'Male'], hue='DRK_YN', palette='Set2', ax=axs[0])
axs[0].set_xlabel('Level of hemoglobin')
axs[0].set_ylabel('Density')
axs[0].set_title('Level of hemoglobin over men')
axs[0].axvline(x=13, linestyle='solid', color='black') # нижний предел гемоглобина у мужчин
axs[0].axvline(x=16, linestyle='solid', color='black') # верхний предел гемоглобина у мужчин

sns.histplot(x='hemoglobin', data=df[df['sex'] == 'Female'], hue='DRK_YN',palette='Set2', ax=axs[1])
axs[1].set_xlabel('Level of hemoglobin')
axs[1].set_ylabel('Density')
axs[1].set_title('Level of hemoglobin over women')
axs[1].axvline(x=12, linestyle='solid', color='black') # нижний предел гемоглобина у женщин
axs[1].axvline(x=16, linestyle='solid', color='black') # верхний предел гемоглобина у женщин


plt.suptitle('Level of hemoglobin over sexes', fontweight='bold', fontsize=18)
plt.show()
```

Из графика можно заметить, что уровень гемоглобина не сильно зависит от того, пьющий человек или нет, так как среди пьющих людей большая концентрация тех, у кого уровень гемоглобина находится в норме. Но все же можно выделить интересную особенность: у пьющих мужчин, если уровень гемоглобина выходит за пределы нормы, то он её превышает, а у женщин наоборот - ниже нормы

```python
sns.countplot(x='SMK_stat_type_cd', hue='urine_protein', data=df, palette='Set1')
plt.xlabel('Smoke or not (1 - never, 2 - quit, 3 - still)')
plt.ylabel('Urine protein')
plt.title('Urine protein level indicators over smoke statuses', fontweight='bold', fontsize=18)
plt.show()
```

График показывает, что уровень белка в моче не зависит от статуса курения человека, так как во всех 3 группах преобладает 1 статус

```python
plt.figure(figsize=(12, 5))
for drk in np.unique(df['DRK_YN']):
  sns.kdeplot(df[df['DRK_YN'] == drk]['waistline'], label=drk)

plt.ylabel('Density')
plt.xlabel('Waistline')
plt.legend(title='Drink or not')
plt.suptitle('Distribution of waistlines over drinkers and non-drinkers', fontweight='bold', fontsize=18)
plt.show()
```

Из графика видно, что у пьющих людей немного больше линия талии  
Теперь проверим гипотезы 5-10, сформулированные перед началом исследования

```python
crosstab = pd.crosstab(df['DRK_YN'], df['hear_left'])
crosstab
```

Все ячейки содержат ожидаемую частоту более 5, так что тест хи-квадрат применим, но интересным наблюдением является то, что отклонение от нормы слуха больше у **непьющих** людей

```python
from scipy.stats import chi2_contingency as chi2
chi2(crosstab)
```

Вывод: гипотеза Н1 принимается на уровне значимости 5% ($pvalue = 0 < 0.05 ⇒$) существует взаимосвязь между фактом того, пьющий человек или нет и его уровнем слуха на левом ухе

```python
crosstab = pd.crosstab(df['SMK_stat_type_cd'], df['hear_left'])
crosstab
```

Все ячейки содержат ожидаемую частоту более 5, так что тест хи-квадрат применим, но вновь интересным наблюдением является то, что отклонение от нормы слуха больше у людей, которые никогда не курили

```python
chi2(crosstab)
```

Вывод: гипотеза Н1 принимается на уровне значимости 5% ($pvalue=1.77 * 10^{-224}< 0.05⇒ $) существует взаимосвязь между статусом курения человека и его уровнем слуха на левом ухе

```python
crosstab = pd.crosstab(df['hear_left'], df['hear_right'])
crosstab
```

Все ячейки содержат ожидаемую частоту более 5, так что тест хи-квадрат применим. Сводная таблица показывает, что почти все наблюдаемые имееют хороший слух на обоих ушах, отклонение хотя бы на одном ухе является редкостью в рамках данного датасета

```python
chi2(crosstab)
```

Вывод: гипотеза Н1 принимается на уровне значимости 5% ($pvalue= 0 < 0.05⇒ $) существует взаимосвязь между уровнем слуха человека на левом и правом ухе

```python
crosstab = pd.crosstab(df['sex'], df['DRK_YN'])
crosstab
```

Все ячейки содержат ожидаемую частоту более 5, так что тест хи-квадрат применим. Сводная таблица показывает, что среди мужчин пьющих примерно в 2 раза больше, чем непьющих, в то время как среди женщин ситуация обратная - пьющих примерно в 2 раза меньше, чем непьющих

```python
crosstab = pd.crosstab(df['sex'], df['SMK_stat_type_cd'])
crosstab
```

```python
print(df.agg({"age": ["mean", "std"], "height": ["mean", "std"],
                                            "weight": ["mean", "std"],
                                            "waistline": ["mean", "std"],
                                            "sight_left": ["mean", "std"],
                                            "sight_right": ["mean", "std"],
                                            "hear_left": ["mean", "std"],
                                            "hear_right": ["mean", "std"],
                                            "SBP": ["mean", "std"],
                                            "DBP": ["mean", "std"],
                                            "BLDS": ["mean", "std"],
                                            "tot_chole": ["mean", "std"],
                                            "hemoglobin": ["mean", "std"]}))
```

В целом, никаких поразительно выбивающихся статистических параметров нет, что говорит нам о том, что данные примерно имеют нормальное распределение. В подтверждение этого проведем тест Колмогорова-Смирнова, и проверим гипотезы 1-4.

```python
from scipy import stats
stats.kstest(df['age'].dropna(), 'norm', args=(df['age'].mean(), df['age'].std()))
```

Н1: распределение возраста наблюдаемых статистически значимо отличается от нормального распределения.

Вывод: поскольку p-value < 0.05, гипотеза H1 принимается

```python
stats.kstest(df['height'].dropna(), 'norm', args=(df['height'].mean(), df['height'].std()))
```

Н1: распределение роста наблюдаемых статистически значимо отличается от нормального распределения.

Вывод: поскольку p-value < 0.05, гипотеза H1 принимается

```python
stats.kstest(df['weight'].dropna(), 'norm', args=(df['weight'].mean(), df['weight'].std()))
```

H1: распределение веса наблюдаемых статистически значимо отличается от нормального распределения.

Вывод: поскольку p-value < 0.05, гипотеза H1 принимается

```python
stats.kstest(df['waistline'].dropna(), 'norm', args=(df['waistline'].mean(), df['waistline'].std()))
```

H1: распределение линии талии наблюдаемых статистически значимо отличается от нормального распределения.

Вывод: поскольку p-value < 0.05, гипотеза H1 принимается

```python
print(f"Тест для sight_left: {stats.kstest(df['sight_left'].dropna(), 'norm', args=(df['sight_left'].mean(), df['sight_left'].std()))}")
print(f"Тест для sight_right: {stats.kstest(df['sight_right'].dropna(), 'norm', args=(df['sight_right'].mean(), df['sight_right'].std()))}")
print(f"Тест для SBP: {stats.kstest(df['SBP'].dropna(), 'norm', args=(df['SBP'].mean(), df['SBP'].std()))}")
print(f"Тест для DBP: {stats.kstest(df['DBP'].dropna(), 'norm', args=(df['DBP'].mean(), df['DBP'].std()))}")
print(f"Тест для BLDS: {stats.kstest(df['BLDS'].dropna(), 'norm', args=(df['BLDS'].mean(), df['BLDS'].std()))}")
print(f"Тест для tot_chole: {stats.kstest(df['tot_chole'].dropna(), 'norm', args=(df['tot_chole'].mean(), df['tot_chole'].std()))}")
print(f"Тест для hemoglobin: {stats.kstest(df['hemoglobin'].dropna(), 'norm', args=(df['hemoglobin'].mean(), df['hemoglobin'].std()))}")
```

По тесту Колмогорова-Смирнова можно увидеть, что ни одна из исследуемых переменных не имеет нормального распределения, поэтому наше предположение, что переменные имеют примерно нормальное распределение, ошибочное.

Теперь разобьем статистику на группу пьющих и непьющих людей и посмотрим, что поменяется.

Добавим также категориальные предикторы в нашу модели, предварительно перекодировав их в набор дамми-переменных

```python
DRK_dummies = pd.get_dummies(df.DRK_YN, prefix='DRK', prefix_sep='_')
DRK_dummies
```

```python
SMK_dummies = pd.get_dummies(df.SMK_stat_type_cd, prefix='SMK', prefix_sep='_')
SMK_dummies
```

```python
df['SMK_stat_type_cd'].value_counts() # SMK_1.0 - референтная группа
```

```python
import statsmodels.api as sm

x = pd.concat([df[['age', 'sight_left', 'sight_right', 'BLDS', 'tot_chole', 'height', 'weight', 'waistline']],
               DRK_dummies['DRK_Y'],
               SMK_dummies[['SMK_2.0', 'SMK_3.0']]],
               axis=1
              ) # SMK_1.0 - референтная группа, DRK_N - референтная группа
y = df['hemoglobin']

x = sm.add_constant(x)

model = sm.OLS(y, x).fit()

print(model.summary())
```

```python
print('R^2 = {:f}'.format(model.rsquared))
print('F-test significance = {:f}'.format(model.f_pvalue))
```

Модель статистически значима, объясняет 33.9% вариации значений зависимой переменной

```python
print(f'Статистически значимые параметры модели: {model.pvalues[model.pvalues < 0.05].index.values}')
```

```python
print(f'Статистически не значимые параметры модели: {model.pvalues[model.pvalues >= 0.05].index.values}')
```

То есть все параметры модели статистически значимые (то есть эти параметры получились именно такими **неслучайно**, а вследствие выявления каких-то закономерностей в данных), кроме переменной age

# Составление модели бинарной регрессии  
Предикторы оставим теми же, что и в случае линейной регрессии, исключив, предиктор DRK_YN, так как это теперь будет целевая переменная. Также добавим в предикторы уровень гемоглобина (hemoglobin), так как он теперь не является целевой переменной (его корреляции с остальными метрическими признаками мы считали, и выяснили, что сильных корреляций нет, так что мы можем его добавить в нашу модель)

# Кластерный анализ

Создадим новый датасет и отнормируем в нем переменные

```python
df1 = df[['age', 'height', 'weight', 'waistline', 'sight_left', 'sight_right', 'SBP', 'DBP', 'BLDS', 'tot_chole', 'hemoglobin', 'urine_protein']].copy()
df1['age'] = (df1['age'] - df1['age'].mean())/df1['age'].std()
df1['height'] = (df1['height'] - df1['height'].mean())/df1['height'].std()
df1['weight'] = (df1['weight'] - df1['weight'].mean())/df1['weight'].std()
df1['waistline'] = (df1['waistline'] - df1['waistline'].mean())/df1['waistline'].std()
df1['sight_left'] = (df1['sight_left'] - df1['sight_left'].mean())/df1['sight_left'].std()
df1['sight_right'] = (df1['sight_right'] - df1['sight_right'].mean())/df1['sight_right'].std()
df1['SBP'] = (df1['SBP'] - df1['SBP'].mean())/df1['SBP'].std()
df1['DBP'] = (df1['DBP'] - df1['DBP'].mean())/df1['DBP'].std()
df1['BLDS'] = (df1['BLDS'] - df1['BLDS'].mean())/df1['BLDS'].std()
df1['tot_chole'] = (df1['tot_chole'] - df1['tot_chole'].mean())/df1['tot_chole'].std()
df1['hemoglobin'] = (df1['hemoglobin'] - df1['hemoglobin'].mean())/df1['hemoglobin'].std()
df1['urine_protein'] = (df1['urine_protein'] - df1['urine_protein'].mean())/df1['urine_protein'].std()
df1.head()
```

Посчитаем корреляции, на основе которых сделаем вывод о том, какие переменные использовать для кластеризации

```python
df1.corr()
```

Выберем для кластеризации следующий набор переменных: age, height, waistline, tot_chole, hemoglobin

```python
df[['age', 'height', 'waistline', 'tot_chole', 'hemoglobin']].corr()
```

Так как датасет очень большой, то построить дендрограммы для него вычислительно невозможно, так что проведем кластерный анализ на произвольных 42 наблюдениях

```python
df1 = df1.sample(n=42)
df1
```

Построим дендрограмму по полученной выборке. Для алгоритма кластеризации используем метод Варда, так как он подходит для небольшого количества объектов и интеллектуально отбирает кластеры, стараясь минимизировать отклонения значений от центра.

```python
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
plt.figure(figsize=(15, 5))
dn = dendrogram(linkage(df1[['age', 'height', 'waistline', 'tot_chole', 'hemoglobin']], 'ward'))
```

Визуально очевидно, что оптимальное число кластеров - 3. Подтвердим это с помощью подсчета индекса Калински-Харабаша для значений от 2 до 9

```python
from sklearn.metrics import calinski_harabasz_score

Z = linkage(df1[['age', 'height', 'waistline', 'tot_chole', 'hemoglobin']], method='ward', metric='euclidean')

for k in range (2, 10):
    labels = fcluster(Z, t=k, criterion='maxclust')
    print('Число кластеров: {}, индекс {}'.format(k, calinski_harabasz_score(df1[['age', 'height', 'waistline', 'tot_chole', 'hemoglobin']], labels)))
```

Наибольшее значение действительно принимается при 3 кластерах, так что это оптимальный размер

```python
labels = fcluster(Z, t=3, criterion='maxclust') #проводим рассчёты для 7-и кластеров
labels
```

```python
df1['culster_labels'] = labels # создаём новую переменную, отражающую принадлежность наблюдения к кластеру
df1
```

```python
import requests
from bs4 import BeautifulSoup
```

Сначала мы решаем считать первые 10 страниц (можно и больше, но программа будет дольше грузиться, она и так 5 минут работает при обработке всех ссылок) с книгами и занести в словарь ссылки на них, чтобы потом каждую страницу с книгой разбирать на данные.  
Так как страницы, со второй, нумеруются с добавлением к адресу: "/page-" и номер страницы — я меняю url при обработке каждой из них.  

```python
url = 'https://knigogid.ru/books'
countPage = 0
links = set()

while countPage != 10:
  r = requests.get(url)
  page = BeautifulSoup(r.text, 'html.parser')
  books = page.find('div', class_='b-items-container genres_books-list').findAll('a', class_='b-item-name')
  for el in books:
    links.add('https://knigogid.ru' + el.get('href'))
  countPage += 1
  url = 'https://knigogid.ru/books' + '/page-' + str(countPage)
```

После создания словаря с ссылками начинаем проходиться по каждой, вынимать данные по условию задачи и создавать массив с требуемой информацией о каждой книге. Мы решили исключать книги, не имеющие каких-либо значений, чтобы не было пропусков в таблице.

```python
dataBooks = []

for link in links:
  r = requests.get(link)

  page = BeautifulSoup(r.text,'html.parser')

  nameBook = page.find('div', class_='b-book-right__top__left').find('h1', class_='b-book-name').text

  author = page.find('div', class_='b-book-user__text').find('a', class_='b-book-user__name').text

  rating = page.find('div', class_='b-book-image__wrap').find('div', class_='b-item-rate').text

  mainGenre = page.find('div', class_='b-book-tags').find('li').text

  language = None
  pages = None
  listPages = page.find('div', class_='b-book-tabs__info').findAll('li')[3].text.split()
  if len(listPages) > 2:
    pages = listPages[2]
  else:
    language = listPages[1]
  if language == None:
    language = page.find('div', class_='b-book-tabs__info').findAll('li')[4].text.split()[1]

  if language == 'ru':
    language = 'Русский'

  isbn = page.find('div', class_='b-book-tabs__info').findAll('li')[2].text.split()[-1]

  countViews = page.find('div', class_='b-book-info__list').find('li').text.split()[-1]

  yearWriting = page.find('div', class_='b-book-tabs__info').findAll('li')[-1].text.split()[-1]

  if yearWriting == 'указан':
    continue

  yearRest = page.find('div', class_='b-book-tabs__info').findAll('li')[-2].text.split()[-1]

  if yearRest == 'указан':
    continue

  if yearRest[-1] == '+':
    yearRest = yearRest[0:-1]

  if pages == None:
    continue

  dataBooks.append([nameBook, author, rating, mainGenre, yearWriting, pages, yearRest, countViews, isbn, language, link])
```

По данным со сайтов составляем свой датафрейм, чтобы потом данные перенести в MS Excel. Также прописываем название каждого столбца, заменяя автоматически сгенерированные индексы.

```python
df = pd.DataFrame(dataBooks)

df = df.rename(columns={0: 'Название_книги', 1: 'Автор', 2: 'Рейтинг_читателей', 3: 'Основной_жанр', 4: 'Год_выпуска', 5: 'Количество_страниц',
                        6: 'Возрастное_ограничение', 7: 'Количество_просмотров_книги', 8: 'ISBN',
                        9: 'Доступный_язык', 10: 'Ссылка_на_книгу'})
df.dtypes
```

Меняем целочисленные числа и десятичные дроби на соответствующие типы.

```python
df['Рейтинг_читателей'] = df['Рейтинг_читателей'].astype(float)
df['Год_выпуска'] = df['Год_выпуска'].astype(int)
df['Количество_страниц'] = df['Количество_страниц'].astype(int)
df['Количество_просмотров_книги'] = df['Количество_просмотров_книги'].astype(int)
df['Возрастное_ограничение'] = df['Возрастное_ограничение'].astype(int)
df
```

Создаем MS Excel файл, с помощью функции to_excel() из библиотеки pandas.
